```python
import numpy as np  
import pandas as pd 
import seaborn as sns
```

```python
df = pd.read_excel("../data/vigitel 2019 com colunas traduzidas.xls", skiprows=[0])
```

```python
# Setar opção de mostrar todas as colunas da tabela
pd.set_option('display.max_columns', None)
df.head()
```

```python
# Substituir nome colunas
df = df.rename(columns={'Unnamed: 3': 'Cidade','Unnamed: 2': 'Ano', 'Unnamed: 1': 'Replica', 'Unnamed: 0': 'Chave'})
```

```python
# Listar colunas da tabela
df.columns.values.tolist()
```

```python
# Deletar colunas que divergem do dataset da hygia
df = df.drop(columns=['qual era o peso quando tinha 20 anos', 'possui habilitação', 'dirige', 'frequência horta crua', 'frequência horta cozida'])
```

```python
# Pegar as 13 primeiras colunas
primeirasTrezeColunas = df.iloc[: , 0 : 13].copy()
primeirasTrezeColunas.head()
```

```python
# media peso em kg no território brasileiro
mediaPeso = primeirasTrezeColunas['peso em kg'].mean() 
print("Média ", mediaPeso)
```

```python
# moda peso em kg no território brasileiro (peso que mais aparece)
modaPeso = primeirasTrezeColunas['peso em kg'].mode() 
print("Moda Peso ", modaPeso[0])
```

```python
# variancia peso em kg no território brasileiro (quão afastados os dados estão da média)
# Quanto maior a variância, mais afastados os dados encontram-se da média
# Por que a variância está tão alta?
varPeso = primeirasTrezeColunas['peso em kg'].var() 
print("Variância Peso ", varPeso)
```

```python
# plotei os pesos no gráfico abaixo e podemos constatar que existem valores absurdos (pesos acima de 300kg)
# logo, temos que tratar os casos 'outliers'
primeirasTrezeColunas['peso em kg'].plot(xlabel='Número de pessoas', ylabel='Peso')
```

```python
# apenas para visualização dos dados: todas as colunas em gráficos
axs = primeirasTrezeColunas.plot.area(figsize=(12, 20), subplots=True)
```

```python
# Boxplot é utilizado para avaliar e comparar o formato, tendência central e variabilidade de distribuições de amostra.
# Serve também para procurar por outliers. 
# O retangulo em azul esta divido, sendo a linha que divide a Mediana
# O último traço à direita estão os dados atípicos (fora do intervalo) 
sns.boxplot(primeirasTrezeColunas['peso em kg'])
```

```python
# Gráfico em histograma
# Cada barra agrupa números em intervalos 
# As barras mais altas mostram que mais dados estão nesse intervalo
sns.distplot(primeirasTrezeColunas['peso em kg']) 
```
